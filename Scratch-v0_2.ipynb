```python
class KalmanFilter:
    def __init__(self, udim, xdim, zdim, cuda=False):
        super().__init__()
        self.udim = udim
        self.xdim = xdim
        self.zdim = zdim
        self.device = 'cuda' if cuda else 'cpu'
        
        self.model = self._model()
        self.guide = self._guide()
    
    def _model(self):
        def model(u, z):
            u, z = u.to(self.device), z.to(self.device)
            # a HalfNormal can be used here as well
            Q = 0.1*torch.ones(self.xdim, device=self.device)
            R = 0.1*torch.ones(self.zdim, device=self.device)
            
            # State transition matrix
            #F = pyro.sample('state_transition', dist.MultivariateNormal(
            #        torch.zeros(self.xdim*self.xdim), 1e-2*torch.eye(self.xdim*self.xdim))).reshape(self.xdim,self.xdim)
            #F = pyro.sample('state_transition', dist.Beta(
            #        2.0*torch.ones(self.xdim*self.xdim, device=self.device),
            #        2.0*torch.ones(self.xdim*self.xdim, device=self.device))).reshape(self.xdim,self.xdim)
            with pyro.plate('plate_state', self.xdim):
                F = pyro.sample('state_transition', dist.Beta(
                        2.0*torch.ones(self.xdim, device=self.device),
                        2.0*torch.ones(self.xdim, device=self.device)))
            #F = torch.diag(F)
            #F = pyro.sample('state_transition', dist.Dirichlet(
            #        0.5*torch.ones(self.xdim*self.xdim, device=self.device))).reshape(self.xdim,self.xdim)
            
            # External input transition matrix
            #G = pyro.sample('input_transition', dist.MultivariateNormal(
            #        torch.zeros(self.xdim*self.udim), 1e-2*torch.eye(self.xdim*self.udim))).reshape(self.xdim,self.udim)
            with pyro.plate('plate_input', self.xdim*self.udim):
                G = pyro.sample('input_transition', dist.Normal(
                        torch.zeros(self.xdim*self.udim, device=self.device),
                        0.1*torch.ones(self.xdim*self.udim, device=self.device))).reshape(self.xdim,self.udim)
            
            # Measurement transition matrix
            #H = pyro.sample('measurement_transition', dist.MultivariateNormal(
            #        torch.zeros(self.zdim*self.xdim), 1e-2*torch.eye(self.zdim*self.xdim))).reshape(self.zdim,self.xdim)
            #H = pyro.sample('measurement_transition', dist.MultivariateNormal(
            #        torch.zeros(self.zdim*self.xdim, device=self.device),
            #        0.2*torch.eye(self.zdim*self.xdim, device=self.device))).reshape(self.zdim,self.xdim)
            with pyro.plate('plate_measurement', self.zdim*self.xdim):
                H = pyro.sample('measurement_transition', dist.HalfNormal(
                        1.0*torch.ones(self.zdim*self.xdim, device=self.device))).reshape(self.zdim,self.xdim)
                #H = pyro.sample('measurement_transition', dist.Normal(
                #        torch.zeros(self.zdim*self.udim), 1e-1*torch.ones(self.zdim*self.udim))).reshape(self.zdim,self.udim)
            with pyro.plate('plate_bias', self.zdim):
                bias = pyro.sample('measurement_bias', dist.Normal(
                        torch.zeros(self.zdim, device=self.device),
                        0.2*torch.ones(self.zdim, device=self.device)))
            
            # Prepare time series variables
            time_steps = len(u)
            X, Z = [], []
            
            # Initialize state
            X0 = torch.zeros(self.xdim, device=self.device)
            
            for t in range(time_steps):
                # Predict next step
                with pyro.plate(f'plate_state_{t}', self.xdim):
                    if t == 0:
                        X += [pyro.sample(f'state_{t}', dist.Normal(F * X0 + G @ u[t,:], Q))]
                    else:
                        X += [pyro.sample(f'state_{t}', dist.Normal(F * X[t-1] + G @ u[t,:], Q))]
                
                # Measurement
                with pyro.plate(f'measurement_{t}', self.zdim):
                    Z += [pyro.sample(f'obs_{t}', dist.Normal(H @ X[t] + bias, R), obs=z[t,:])]
                    #Z += [pyro.sample(f'obs_{t}', dist.MultivariateNormal(H @ u[t,:], R), obs=z[t,:])]
        
        return model
    
    def _guide(self):
        #guide = pyro.infer.autoguide.AutoDiagonalNormal(self.model)
        guide = pyro.infer.autoguide.AutoDelta(self.model)
        
        #pyro.param('auto_state_transition', torch.zeros(self.xdim*self.xdim), constraint=tdist.constraints.positive)
        #pyro.param('auto_measurement_transition', torch.zeros(self.zdim*self.xdim), constraint=tdist.constraints.positive)
        
        return guide
    
    def filter(self, x):
        raise NotImplementedError("Forward filtering not yet implemented!")
```
